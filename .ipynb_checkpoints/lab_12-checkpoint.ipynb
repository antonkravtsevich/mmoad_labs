```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn import metrics
from sklearn import preprocessing
from sklearn import tree
```

```python
df = pd.read_table('data/Vote.txt')
```

```python
len(df)
```

```python
df.head()
```

```python
# Удаляем ненужный столбец с кодом
df = df.drop(columns=['Код'])
```

```python
# Кодируем даннные, так как DecisionTreeClassifier не работает с текстовыми данными 
# используем one-hot encoding

values_set = ['да', 'нет', 'воздержался', 'республиканец', 'демократ']
encoder = preprocessing.LabelEncoder()
encoder.fit(values_set)

for column_name in df.columns:
    df[column_name] = encoder.transform(df[column_name])
```

```python
df.head()
```

```python
# Отделяем признаки от основного датасета

y = df['Класс'] # Целевая переменная
X = df.drop(columns=['Класс']) # Набор признаков
```

```python
# Разделяем данные для тренировки и тестовfeatures_names

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)
```

```python
# Создаем классификатор
clf = DecisionTreeClassifier()
```

```python
# обучаем классификатор на тренировочных данных
clf = clf.fit(X_train,y_train)
```

```python
# предсказываем значения для тестовых данных

y_pred = clf.predict(X_test)
```

```python
# измеряем точность предсказания

print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```
