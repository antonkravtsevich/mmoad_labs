# 1. Парциальная обработка

```python
df = pd.read_csv('./data/Trade.txt')
```

```python
df.head()
```

```python
ax = df['Количество'].plot()
ax.set_xticklabels(df['Дата (Год + Месяц)'])
```

```python
df['Количество'][df['Количество'].between(df['Количество'].quantile(.95), df['Количество'].quantile(1.0))] = df['Количество'].quantile(.95)
```

```python
ax = df['Количество'].plot()
ax.set_xticklabels(df['Дата (Год + Месяц)'])
```

# 2. Спектральная обработка

```python
from scipy import signal
```

```python
N  = 3    # Filter order
Wn = 0.25 # Cutoff frequency
B, A = signal.butter(N, Wn, output='ba')
df['Количество_2'] = signal.filtfilt(B,A, df['Количество'])
df['Количество_2'].plot()
```

# 3. Факторный анализ

```python
df = pd.read_table('data/Anketa1.txt')
```

```python
df.head()
```

```python
X = df[[
    'Личный доход в месяц после налогооблажения',
    'Сумма кредита, руб#',
    'Стаж работы, лет',
    'Рыночная стоимость автомобиля, руб#',
    'Рыночная стоимость недвижимости, руб#'
]]
```

```python
X.info()
```

```python
# изменяем тип данных для стажа работы
X['Стаж работы, лет'] = X['Стаж работы, лет'].map(lambda x: x.replace(',', '.'))
```

```python
X['Стаж работы, лет'] = X['Стаж работы, лет'].astype(float)
```

```python
X.info()
```

```python
from factor_analyzer import FactorAnalyzer
```

```python
# Выбираем количество факторов
```

```python
fa = FactorAnalyzer()
```

```python
fa = fa.fit(X)
```

```python
ev, v = fa.get_eigenvalues()
```

```python
ev
```

```python
# только три собственных значения примерно равны 1, значит можно использовать только три фактора
```

```python
fa = FactorAnalyzer().fit(X, 3)
```

```python
X_factors_values = fa.transform(X)
```

```python
# Добавляем факторы в основной набор данных

df['Фактор 1'] = X_factors_values.T[0]
df['Фактор 2'] = X_factors_values.T[1]
df['Фактор 3'] = X_factors_values.T[2]
```

```python
df
```
