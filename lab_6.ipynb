2. Отсортируйте этот набор данных по следующим полям в порядке
возрастания: Срок ссуды, Размер ссуды, Количество иждивенцев.

```python
df = df.sort_values(['Срок ссуды, мес', 'Размер ссуды, руб', 'Количество иждивенцев'])
df
```

3. Сделайте следующую замену (после Сортировки) в поле Семейное
положение: значение Да измените на Женат/замужем, Нет – на
Холост/Не замужем.

```python
df = df.replace({'Семейное положение': {'Да': 'Женат/замужем', 'Нет': 'Холост/Не замужем'}})
df
```

4. Сделайте следующую замену (после предыдущего узла Замена
данных) в поле Количество иждивенцев: значение 0 – на Нет, 1 – без
изменений, 2 и 3 – 2 и более. Используйте два способа –
непосредственным вводом в мастере обработки и через файл таблицы
соответствий. Файл подстановок предварительно создайте в любом
текстовом редакторе, например, в Блокноте.

```python
def change_value(value):
    if value == 0:
        return 'Нет'
    if value == 1:
        return 'Без изменений'
    if value >= 2:
        return '2 и более'
```

```python
df['Количество иждивенцев'] = df['Количество иждивенцев'].map(lambda x: change_value(x))
df
```

5. Старое поле Количество иждивенцев удалите из набора данных, а
новое поле Количество иждивенцев_REPLACE переименуйте в
Иждивенцы.

```python
df = df.rename(columns={'Количество иждивенцев': 'Инждивенцы'})
df
```

8. Продолжите фильтровать набор данных, полученный в п. 6. Наложите
следующий фильтр, в который попадают все записи, удовлетворяющие
условиям a либо условиям b:

a. Размер ссуды – от 2000 до 5000, Цель ссуды – Покупка товара.

b. Цель ссуды – Иное.

```python
df = df[df['Инждивенцы']!='Нет']
```

```python
df.columns
```

```python
a_conditions = (df['Размер ссуды, руб'] >= 2000) & (df['Размер ссуды, руб'] <= 5000) & (df['Цель ссуды'] == 'Покупка товара')
b_conditions = df['Цель ссуды'] == 'Иное'
```

```python
df[(a_conditions | b_conditions)]
```

9. Сколько записей прошло через фильтр?

```python
print(f"Количество записей: {len(df[(a_conditions | b_conditions)])}")
```
